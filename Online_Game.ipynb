# filtering 

```python
# Q) Filter games with Global Sales greater than 1 million units
high_gb_sales = df[df['Global_Sales']>1]
high_gb_sales.reset_index(drop = True,inplace = True)
high_gb_sales.head()
```

```python
# Q) Selecting games that sold more than 1 million units in North America\
high_na_sales = df[df['NA_Sales']>1]
high_na_sales.reset_index(drop = True,inplace=True)
high_na_sales.head()
```

```python
# Q)Selecting the top 10 ranked games
df.iloc[:10]
```

```python
#  Q) Filtering for games that sold more in Japan than in North America
df[df['JP_Sales']>df['NA_Sales']]
```

```python
# Q) Filtering for games available on a PS4 platform 
PS4 = df[df['Platform']=='PS4']
PS4.reset_index(drop = True,inplace =True)
PS4.head()
```

# sort()

```python
# Q)  Sort games by Global Sales in descending order
gb_sales = df.sort_values(by = 'Global_Sales',ascending = False)
gb_sales.head()
```

```python
# Q) Sort games by Genre and then by Global Sales within each Genre in descending order
df[['Global_Sales','Genre']].sort_values(by = 'Genre',ascending = False)
```

```python
# Q) Sort games by Publisher, and then by Year in ascending order
df[['Publisher','Year']].sort_values(by = 'Year',ascending = True)
```

```python
# Q) Sort games by JP Sales in descending order to find the top-selling games in Japan
jp_sales = df.sort_values(by = 'JP_Sales',ascending = False)
jp_sales.reset_index(drop = True,inplace = True)
jp_sales.head()
```

```python
# Q) Sort games by Platform, and within each Platform, sort by Global Sales in descending order
df[['Platform','Global_Sales']].sort_values(by = 'Global_Sales',ascending = False)
```

# isin()

```python
# Q) Find games from the genres "Action", "Adventure", or "RPG"
specific_genre = df[df['Genre'].isin(['Action','Avdeventure','RPG'])]
specific_genre.reset_index(drop = True,inplace = True)
specific_genre.head()
```

```python
# Q) Filter games released in the years 2010, 2015, or 2020
specific_year = df[df['Year'].isin([2010,2015,2020])]
specific_year.reset_index(drop = True,inplace = True)
specific_year.head()
```

```python
# Q) create a new DataFrame that excludes all rows where the 'Platform' column contains any of the values 'PS', 'PS2', 'PS3', or 'PS4'
Excluded_codes = ['PS', 'PS2', 'PS3','PS4']
df[~df['Platform'].isin(['Excluded_codes'])]
```

# pivot_table()

```python
# Q) create a pivot table that shows the sum of global sales for each genre across different years?
df.pivot_table(index = 'Year',columns = 'Genre',values = 'Global_Sales',aggfunc = sum)
```

# Groupby

```python
# Q) group the dataset by both game name and year, and then calculate the maximum total sales ratio and minimum global sales for each game in each year?
df.groupby(['Name','Year']).agg({'Total_Sales_Ratio':max,'Global_Sales':min})
```

```python
# Q) group the dataset by game name and platform, and then calculate the maximum, minimum, mean, and total North American sales for each combination?
df.groupby(['Name','Platform'])['NA_Sales'].agg(['max','min','mean','sum'])
```
